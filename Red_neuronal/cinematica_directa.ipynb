```python
dataf = pd.read_csv("full_data.csv", header=1, decimal=",", sep=";")
```

```python
dataf
```

```python
dataf = dataf.iloc[:1399] # Me quedo con los 1400 primeros puntos
# Shuffle the data
dataf = dataf.sample(frac=1).reset_index(drop=True)
dataf
```

```python
Y = dataf[["Extremo X", "Extremo Y"]].to_numpy()
```

```python
Y
```

```python
X = dataf.to_numpy()[:, 6::2]
```

Nos quedamos con todas las filas y despreciamos las seis primeras columnas y luego cogemos las columnas restantes una sí una no, para quedarnos con los motores impares.
Y son las posiciones (x,y) y X son los motores impares

```python
X.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, make_scorer
from sklearn.model_selection import GridSearchCV
```

He distribuido los valores de X y de Y entre -1 y 1 haciendo una transformación lineal. Es solo porque los algoritmos se suelen llevar mejor con números pequeños y parecidos entre sí, nada más.
También me he cargado las posiciones Z, no sé si con buen o con mal criterio. Me ha parecido con un vistazo rápido que el robot se mueve en el plano (x,y).
También elijo entrenar solo los motores impares, los pares se pueden obtener a partir de ellos.

```python
y_scaler = MinMaxScaler((-1,1)).fit(Y)
Y = y_scaler.transform(Y)
```

```python
y_scaler.data_min_, y_scaler.data_max_
```

```python
import joblib
joblib.dump(y_scaler, "yscaler.joblib")
```

```python
scaler = MinMaxScaler((-1,1))
```

```python
regressor = MLPRegressor(
    verbose=False,
    random_state=42,
    activation = "relu",
    early_stopping = True,
    max_iter=1000,
)
```

random state es la semilla del generador de números aleatorios. Solo sirve para que cada vez que ejecutes el script te salga lo mismo. 
early stopping True significa tener un mini test set para parar de entrenar antes de hacer overfitting.
max_iter=600, significa que cada elemento del traning set se le ha enseñado a la red un máximo de 600 veces.
optimizador adam. Recordar citarlo en el TFM.

```python
model = Pipeline(steps =[
    ("scaler", scaler),
    ("nn", regressor),
])
```

```python
param_grid = {"nn__hidden_layer_sizes":[(50,50),(100,100), (150,150), (50,),(100,), (150,)],
             "nn__learning_rate_init": [0.0001,0.0005,0.001,0.005,0.01,0.05],
             "nn__alpha": [0.0001,0.0005,0.001,0.005,0.01,0.05],
            }
```

alpha es el peso de la regularizacion L2, si es pequeño permite overfitting y si es grande produce underfitting. Si es grande fuerza a que los parámetros (W1, W2, W3...) sean demasiado pequeños, si es 0 los parametros pueden ser arbitrariamente grandes.

```python
cv_1layer = GridSearchCV(model, param_grid, scoring=make_scorer(r2_score), verbose=1, n_jobs=4, cv=5)
```

cv = 3 signifca que he hecho cross validation con tres particiones en vez de con 5, porque tardaba la vida. 
n_jobs=4 significa que he usado las 4 CPUs que tiene mi ordenador para entrenar
scoring=r2, significa que he elegido como mejor red la que tiene mejor coeficiente de correlacion entre los y_test y los y_predichos

```python
cv_1layer.fit(X, Y)
```

```python
cv_1layer.best_score_
```
