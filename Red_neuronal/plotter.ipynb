```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
dataf = pd.read_csv("full_data.csv", header=1, decimal=",", sep=";")
```

```python
dataf
```

```python
fig, ax = plt.subplots((1), figsize=(5,5))
plt.plot(dataf["Extremo X"]*100, dataf["Extremo Y"]*100, "o", alpha=0.5)

plt.plot(np.linspace(-2.2, 2.2, 100), -2.26*np.ones(100), "k-")
plt.plot(np.linspace(-2.2, 2.2, 100), 2.26*np.ones(100), "k-")
plt.plot(-2.26*np.ones(100),np.linspace(-2.2, 2.2, 100), "k-")
plt.plot(2.26*np.ones(100),np.linspace(-2.2, 2.2, 100), "k-")
plt.xlabel("x(cm)")
plt.ylabel("y(cm)");
```
